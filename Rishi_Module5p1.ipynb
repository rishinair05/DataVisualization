# Matplotlib Library

In this module you will be learning how to graph data using matplotlib library. We will be using only a small sction of matplotlib for our course which is under pyplot. Using pyplot, we will learn how to make bar graphs, line charts and pie charts. The link to the matplotlib.pyplot documentation is [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot) for reference. Let us begin

```python
# plt is the common short form used for matplotlib.pyplot
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. Line Plots

We will start off with the most basic plots matplotlib is used for, ie, line plots. For line plots, you are basically comparing two lists of numbers on a cartesian plane. Try to follow the next pieces of code to familiarize yourself with the library. 



```python
# We will be using this very simple data that contains the first 11 whole numbers, 
# their squares and their cubes
numbers = {0: [0, 0], 1: [1, 1], 2: [4, 8], 3: [9, 27], 4: [16, 64], 5: [25, 125],
        6: [36, 216], 7: [49, 343], 8: [64, 512], 9: [81, 729], 10: [100, 1000]}
data = pd.DataFrame(data= numbers)
data
```

```python
# We will be plotting the squares and cubes against the original numbers
# To do so we first of all get our x-axis and y-axis values

plt_x = data.columns
sq_y = data.iloc[0]
cube_y = data.iloc[1]

# plotting commands

#plt.plot(x_axis values, y-axis values)
plt.plot(plt_x, sq_y)
plt.plot(plt_x, cube_y)

# Note how both the plots are on the same graph. We can change that!

# Using "plt.figure()" creates literally a new figure on which everything will
# be plotted. plt.figure can be used with any graph, not just line graphs
plt.figure()

# plotting the numbers against themselves
plt.plot(plt_x, plt_x)

# You can even name your figures when you make a new one as:
# plt.figure(num= )
# You can fill the num parameter with either an int or a string. If you use a 
# string, it will automatically assign a new int starting from 1 (if 1 has not been used)
```

```python
# Customizing the Graphs

plt.plot(plt_x, sq_y, '.r--')
plt.plot(plt_x, cube_y, '*k-.')
plt.plot(plt_x, plt_x, '+g:')
# https://matplotlib.org/stable/api/markers_api.html 
# Link to check out all other marker symbols


# Adding titles and labels for axes
plt.xlabel('x')
plt.ylabel('y')
plt.title('Practice Graph')

# Adding a legend
names = ['y = x^2', 'y = x^3', 'y = x']
plt.legend(labels = names)

# Adding a grid
plt.grid()
```

```python
# A Different Way to do the Same Thing

# Customizing the Graphs

plt.plot(plt_x, sq_y, marker= '.', color = '#28CBE3', linestyle= '--', linewidth= 3, markersize= 10, label= 'Squares')
plt.plot(plt_x, cube_y, '*k-.', label= 'Cubes')
plt.plot(plt_x, plt_x, '+g:', label= 'Original')
# https://matplotlib.org/stable/api/markers_api.html 
# Link to check out all other marker symbols
# You can use hex codes for colours as well as the colour shotcuts known by matplotlib

# Adding titles and labels for axes
plt.xlabel('x')
plt.ylabel('y')
plt.title('Practice Graph')

# Adding a legend
plt.legend()

# Adding a grid
plt.grid()
```

```python
# Using premade styles available 

# To see all the possible styles
print(plt.style.available)

# Use one of the styles 
plt.style.use('ggplot')

# Go about doing your normal plots (you can set your own colors as well with these styles)
plt.plot(plt_x, sq_y)
plt.plot(plt_x, cube_y)
plt.plot(plt_x, plt_x)
```

## 2. Bar Plots

The second type of charts that you will probably be using are bar charts. Let's get started

```python
# To start us off with syntaxes, we will use the previous example only

plt.style.use('classic')

plt.bar(plt_x, sq_y, color = '#28CBE3', label= 'Squares')
# plt.plot(plt_x, cube_y, '*k-.', label= 'Cubes')
plt.plot(plt_x, plt_x, '+g:', label= 'Original', linewidth= '4', markersize= 15)
# https://matplotlib.org/stable/api/markers_api.html 
# Link to check out all other marker symbols
# You can use hex codes for colours as well as the colour shotcuts known by matplotlib

# Adding titles and labels for axes
plt.xlabel('x')
plt.ylabel('y')
plt.title('Practice Graph')

# Adding a legend
plt.legend()

# Adding a grid
plt.grid()
```

```python
# For the next part, we will need another library called numpy
import numpy as np
```

```python
# Plotting Multiple bar graphs Together

plt.style.use('seaborn')

# If we use plt.bar on every plot, they will overlap
plt.bar(plt_x, sq_y, color = '#28CBE3', label= 'Squares')
plt.bar(plt_x, cube_y, label= 'Cubes')
plt.bar(plt_x, plt_x, label= 'Original')
plt.legend(loc= 'upper left')
# You cannot even see the bars at the bottom

# One way to deal with it is to offset the x values

# Creating a numpy array starting from 0 till 10
bar_x = np.arange(11)
print(bar_x)

# Taking a custom width
width = 0.25

# Plotting a new bar graph
plt.figure()
plt.bar(bar_x - width, sq_y, width= width, color = '#28CBE3', label= 'Squares')
plt.bar(bar_x, cube_y, width= width,  label= 'Cubes')
plt.bar(bar_x + width, plt_x, width= width, label= 'Original')

plt.legend(loc= 'upper left')

# Setting custom x axis
names = ['Zero', 'One', 'Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten']
plt.xticks(ticks= bar_x, labels= names)
```

```python
# Finally, you can make custom bar charts very easily as well

plt.style.use('default')

plt.barh(plt_x, sq_y, label= 'Squares')
```

## 3. Pie Charts

Final type of graph we will learn today is Pie Chart. Everybody loves pies, even in presentations so having a good idea of how to make them with python can go a long way

```python
plt.style.use('fivethirtyeight')

# Simple example

# The sum of the slices need not be 100
slices = [10, 25, 20, 20, 15, 10]
labels = [10, 25, 20, 20, 15, 10]

plt.pie(slices, labels= labels)

plt.title('Our Pie Chart')
```

```python
# Adding more formatting to the pie chart

slices = [10, 25, 20, 20, 15, 10]
labels = [10, 25, 20, 20, 15, 10]

# Custom Colours (You can also use hex code)
colours = ['red', 'blue', 'green', 'yellow', 'pink', 'white']

# Exploding Slices
explode = [0, 0, 0, 0.1, 0, 0]

plt.pie(slices, labels= labels, wedgeprops= {'edgecolor':'black'}, colors= colours,
        explode= explode, shadow= True)

plt.title('Our Pie Chart')

# New Figure for different customizations
plt.figure()

plt.pie(slices, labels= labels, wedgeprops= {'edgecolor':'black'}, colors= colours,
        explode= explode, shadow= True, startangle= 90, autopct= '%.1f%%')
```
