In Python (and most programming in general), you start counting lists from the number 0. The first element in a list is “number 0”, the second is “number 1”, etc.

As a result, when you want to get single elements out of a list, you can ask a list for that number element:

` >>> a = [5, 10, 15, 20, 25]`

`>>> a[3]`

`20`

`>>> a[0]`

`5`

There is also a convenient way to get sublists between two indices:

  `>>> a = [5, 10, 15, 20, 25, 30, 35, 40]`

  `>>> a[1:4]`

  `[10, 15, 20]`

  `>>> a[6:]`

  `[35, 40]`

  `>>> a[:-1]`

  `[5, 10, 15, 20, 25, 30, 35]`

Ask the user for a string and print out whether this string is a palindrome or not. (A palindrome is a string that reads the same forwards and backwards).


````python
#### IMPLEMENT YOUR YOUR CODE HERE ####
# Use input() function to get get a word
# Hint: try slicing the word and stepping backwords, for example: word[::-1]
word = list(input("Enter a word: "))

for letter in range(len(word) -1, -1, -1):
  print(word[letter])

if list(word) == list(reversed(word)):
  print("The input word is a palindrome")

else:
  print("The input word is not a palindrome")
````

Ask the user for a string and print out whether the string is a pangram or not. (Pangrams are words or sentences that contain every letter of the alphabet atleast once)

````python
# IMPLEMENT YOUR CODE HERE #
word = input("Enter a word: ")

if len(set('abcdefghijklmnopqrstuvwxyz') - set(word.lower())) == 0 :
  print("The word is a pangram.")

else:
  print("The word is not a pangram.")
````
