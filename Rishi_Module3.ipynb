*Get some real data*

Our data for the entire course will always be stored in these tables or charts or python objects called `dataframes df`. - The `load_dataset( )` function gives us a dataframe, which we assign the variable `df` which stands for dataframe. 



````
df = sns.load_dataset(name = 'iris')
````

*   We "load" the data into a variable `df` (stands for 'dataframe')

````
df
````

As we inspect `df` we see the first 5 and the last 5 rows, but see that (at the bottom) there will be 150 rows and 5 columns. Dataframes in the future will be much larger!

In data science we always want to have each **row** be an **observation**. In this case a unique flower and its measurements. The ***columns*** then becomes ***features***. So *petal_length* column/feature (4th from the left) tells us the petal length of all 150 flowers. 

> **Remember: Rows are observations, columns are features!!!**

What if we don't want to see the entire dataframe, but rather just one feature (column of it) of it? We now learn about indexing: 


In python, we have `lists`, `strings` and `dataframes` used to *STORE DATA* such as numbers and letters. If we want to go into these storage systems and retrieve some of the data, we index using `[ ]` <- square brackets. 

Here, we are retrieving the column (aka feature) of `'petal_length' ` make sure to pass it in as a **string** into the `[ ]`. If you don't use a string it won't work (can you think of why?)

````
df['petal_length'] 
````

> **RT: Use square brackets to retrive certain values from various data structures.** In the programming world this is called **'indexing'** Indexing one column of a dataframe gives us a series, which we can plot later!




## Pinpointing DataFrame Locations
You can pinpoint a specific value of a DataFrame by using the `df.loc[]` method.
This method takes in a `list of two` as arguments: the row and the column.

So for example, if I wanted to see what the `petal_length` of the flower in `Row 2`, then I would use the following code:


````
df.loc[2,'petal_length'] = 1.3
````

You can see in the above example that the petal length of the flower in row 2 (the third row in order) is indeed 1.3

````
## OPTIONAL EXERCISE: Determine the following values:
## The sepal length of the 56th row
## The petal width of the 82nd row
## The species of the 101st row
x = df.loc[56, "petal_length"]
y = df.loc[82, "petal_width"]
z = df.loc[101, "species"]
print(x)
print(y)
print(z)
````

## `For Loop` through a DataFrame
There are functions within the pandas module that allows you to loop through each row in a dataframe. This loop is slightly different than the loops you learned, because there are **two** indexing names: `index` and `row`.

In short, `index` gives you the row number you are working on, starting from `0`. The loop will do its code, then move on to the next row number, or the next index of `1`, and so on.

The `row` is the meat of the for loop. It returns a `Series` of that specific row. For example, when index is `0`, the corresponding row is:
`5.1, 3.5, 1.4, 0.2, 'setosa'`

````
df
````

````
for index, row in df.iterrows():  # Don't forget the brackets!
  df['sepal_length'][index] = row['sepal_length'] + 1

df
````

Remember what we learned about indexing? Well, that can be applied here, too. Using a `for loop`, we can use indexing to print out everything in one column by using the square brackets and the name of the column. Let's print out every ten entries in the `sepal_length` column.

````
for index, row in df.iterrows():
  if index % 10 == 0:
    print(row['sepal_length'])
````

Let's check that it indeed did that:

````
df.loc[10,'sepal_length']
````

# 1.3 More Indexing (MASKING)
*What if we only want to look a rows which satisfy a condition?*

For example only the large flowers. Or only the virginica flowers. Could we ask for a few particular rows of the `dataframe df`?

Just like `df['petal_length']` is a *Series*, `df['petal_length'] > 5` is also a series of `True` and `False` - True and False as a data type are called *booleans* 

````
df['petal_length'] > 5
````

Having a *Series* of just `True` and `False` might not seem very useful... but lets assign it to **a new variable name** called `mask`

````
mask = df['petal_length'] > 5
````

````
mask 
````

The key is being able to **INDEX** with this mask (a series of booleans) on top of the original dataframe `df`. Instead of getting all 150 rows **we will only get the rows of the dataframe where** `mask` **has a value of** `True` 





````
df[mask]
````

>Essentially, we are only looking at the flowers in the dataset where the petal length is greater than 5. 
We have another dataframe here, lets assign it to the variable `df_large`. 
Note that 


```
mask = df['petal_length'] > 5
df[mask]
```
***is the same as***
```
df[df['petal_length'] > 5]
```
*We introduced it the first way to make things easier to understand, avoiding the double* `[[ ]]`


````
df_large = df[df['petal_length'] > 5]
````

And we inspect: 

````
df_large
````

Looking at `df_large` again, we can see that all but one of the flowers within are of the `species` `virginica`. This is our first insight in the data, larger flowers (based on `petal_length`) are more likely to be `virginica`



Of course, we want to investigate other species. Lets try to find all the *setosa* flowers. Since the column `'species'` is not a numerical value we can't use >, < >= <= etc to make a **comparison** (You need a comparison to get a `True` or `False`).

>  A comparison for 'equals to' is == not =. **Do NOT use = (single equals sign)**, =  is for assigning new variables. **USE `==` properly**

````
df['species'] == 'setosa'
````

Rather than assigning the above boolean series to another mask variable. I will directly put it directly into the index.



````
df_setosa = df[df['species'] == 'setosa']
````

 A good way to read the above line is: 


> set `df_setosa` equal to the `df` where `df['species']` is exactly equal to `setosa`


````
df_setosa
````

We now have 2(two!) new dataframes in addition to `df` which we started with. `df_large` and `df_setosa`, make some more in the following (mandatory) exercise.
